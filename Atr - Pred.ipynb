```python
# import numpy as np
import pandas as pd
```

```python
d_f = pd.read_csv('/content/Book-01.csv')
```

```python
d_f.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l_e = LabelEncoder()
```

```python
# encoded Attrition
e_atr = l_e.fit_transform(d_f['Attrition'])
```

```python
print(e_atr)
```

```python
X = d_f[['Years At Company','Work Life Balance','Age']]
```

```python
y = e_atr
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
log_reg.fit(X_train,y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
cnf_mtx = metrics.confusion_matrix(y_test,y_pred)
```

```python
print(cnf_mtx)
```

```python
import matplotlib.pyplot as pp
import seaborn as sb
%matplotlib inline
```

```python
sb.heatmap(pd.DataFrame(cnf_mtx))
pp.title('Confusion Matrix',y=1.1)
pp.ylabel('Actual Label')
pp.xlabel('Predicted Label')
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```
