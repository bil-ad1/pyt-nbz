```python
d__f = d_f.drop(['Employee ID','Birth_Year'],axis=1)
```

```python
d__f.describe()
```

```python
X1 = d__f[['Age','Rating','Tenure']]
Y1 = d__f['Salary']
```

```python
from sklearn.linear_model import LinearRegression
mod1 = LinearRegression()
```

```python
mod1.fit(X1,Y1)
```

```python
p_sal = mod1.predict([[25,2,4]])
```

```python
print(p_sal)
```

```python
print(mod1.coef_)
```

```python
print(mod1.intercept_)
```

```python
r_sq = mod1.score(X1,Y1)
```

```python
print(r_sq)
```
