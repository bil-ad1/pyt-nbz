```python
import matplotlib.pyplot as pp
import seaborn as sb
```

```python
d_f = pd.read_csv('/content/drive/MyDrive/d.sets/Book-s1oc.csv')
```

```python
print("Rows    :",d_f.shape[0])
print("Columns :",d_f.shape[1])
```

```python
d_f.info()
```

```python
print("\tTOTAL EMPLOYEES : ",d_f.employees.sum())
```

```python
pd.pivot_table(d_f, index = 'state',values= 'employees',aggfunc=['sum'])
```

```python
pd.pivot_table(d_f, index = 'state',values= 'employees',aggfunc=['sum']).plot(kind='barh',title="State wise employees")
```

```python
pd.pivot_table(d_f, index= 'region',values='employees',aggfunc=['sum'])
```

```python
pp.title('Region wise Employees')
pp.pie([20309,23402,13692,26879],labels=['Central','East','South','West'])
```

```python
pd.pivot_table(d_f,index='store_size',values='employees',aggfunc=['sum'])
```

```python
d_f.groupby('store_size')['employees'].sum().plot(kind='pie',title='Employees by Store Size')
```

```python
pd.pivot_table(d_f,index=['store_size','region'],values='employees',aggfunc=['sum'])
```

```python
sb.countplot(y='store_size',hue='region',data=d_f)
```

```python
Cal = d_f[d_f.state=='California']
```

```python
pd.pivot_table(Cal,index='city',values='employees',aggfunc=['sum'])
```
